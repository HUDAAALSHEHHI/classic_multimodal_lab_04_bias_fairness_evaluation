```python
text_to_speak = "The results look promising, but I am worried about the bias in the dataset."
tts = gTTS(text=text_to_speak, lang="en")
tts.save("sample.mp3")

# Convert MP3 to WAV for robust ASR decoding
subprocess.run(["ffmpeg", "-y", "-loglevel", "quiet", "-i", "sample.mp3", "sample.wav"], check=True)

print("TTS text:", text_to_speak)
```

```python
asr = pipeline("automatic-speech-recognition", model="openai/whisper-tiny.en")
asr_out = asr("sample.wav")
transcript = asr_out["text"]
print("Transcript:", transcript)
```
